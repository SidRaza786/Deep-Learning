```python
# importing all libraries
import tensorflow as tf
from tensorflow import keras
from keras_preprocessing.image import load_img, array_to_img, img_to_array
from keras.applications.resnet import ResNet50, preprocess_input, decode_predictions
```

```python
import numpy as np
```

## **ResNet50**

```python
model = ResNet50(weights = "imagenet")
```

```python
img_path = "/content/cat.jpg"
```

```python
image = load_img(img_path, target_size = (224, 224))
image
```

```python
# now, convert image into array
img_arr = img_to_array(image)
img_arr.shape
```

```python
img_arr = np.array(img_arr).reshape(-1, 224,224,3)
img_arr.shape
```

```python
#img_arr = np.expand_dims(img_arr, axis = 0)
#img_arr.shape
```

```python
#img_arr = preprocess_input(img_arr)
#img_arr.shape
```

```python
pred = model.predict(img_arr)
print("Predicted", decode_predictions(pred, top = 3)[0])
```

```python
print("Predicted", decode_predictions(pred, top = 1)[0])
```
